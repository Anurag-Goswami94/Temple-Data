# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Files

```python
temple_df= pd.read_csv("NewAncientTemples.csv",  encoding='ISO-8859-1') ## As it was showing ('utf-8' codec) error so we used encoding
temple_df
```

```python
temple_df.describe(include="all")
```

```python
temple_df.info()
```

```python
temple_df.isnull().sum()
```

```python
temple_df.isnull()
```

```python
new_temple= temple_df['Temple_Name'].str.split(',', expand=True)
new_temple
```

```python
temple_df['Temple_Names']= new_temple[0]
temple_df['City']= new_temple[1]
temple_df['State']= new_temple[2]
temple_df
```

```python
new_temple_df= temple_df[['Temple_Names', "City", 'State', 'Coordinates','Distance_From_Mumbai(Km)', 'Distance_From_NewDelhi(Km)','Distance_From_Chennai(Km)', 'Distance_From_Kolkata(Km)']]
new_temple_df
```

```python
new_temple_df['State'].fillna(value='Orissa', inplace=True)
new_temple_df['City'].replace(' Orissa', "Konark", inplace=True)
new_temple_df
```

```python
new_temple_df.info()
```

```python
new_temple_df['State'] = new_temple_df['State'].apply(lambda x: x.strip())
new_temple_df
```

#### Which state has the highest number of temple

```python
new_temple_df['State'].value_counts()
```

```python
plot_state= new_temple_df['State'].value_counts().plot.bar(x= 'State')
plot_state
```

#### Which temple is far-away from your home town (Kolkata) and which one is close to home town

```python
kol_ds= new_temple_df[['Temple_Names','State',"Distance_From_Kolkata(Km)"]]
kol_ds
```

```python
kol_ds.plot.box(x='State') ## To check outliers 
```

```python
kol_ds.describe()
```

```python
kol_drop= kol_ds.drop(index=[17,49], axis=0)
kol_drop
```

```python
kolkata_group= kol_drop.groupby('State')
kolkata_group
```

```python
max_kol=kolkata_group.max()
max_kol
```

```python
far_kol= max_kol.sort_values('Distance_From_Kolkata(Km)', ascending= False).head(3).plot.bar(x="Temple_Names")
far_kol
```

```python
min_kol=kolkata_group.min()
min_kol
```

```python
near_kol= min_kol.sort_values('Distance_From_Kolkata(Km)', ascending=True)
near_kol
```

```python
plt_near_kol= near_kol[1:4].plot.bar(x="Temple_Names")
plt_near_kol
```

####   Top 5 temples nearest to each metro city (Mumbai, Delhi, Chennai, Kolkata).

```python
drop_mumbai= top_mumbai.drop(index=29,axis=0)
drop_mumbai
```

```python
mum_group= drop_mumbai.groupby('State')
mum_group
```

```python
mumbai_min= mum_group.min()
mumbai_min
```

```python
near_mumbai= mumbai_min.sort_values('Distance_From_Mumbai(Km)',ascending= True)
near_mumbai
```
